# Analyze the screen time of the user to find relationships between the usage of the smartphone and factors like notifications and apps used by the user.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("https://raw.githubusercontent.com/AldataSci/ScreenTimeAnalysis/main/Screentime-App-Details.csv")
data.head(15)
```

```python
print(data.head())
```

```python
## Here we can see that there is a higher usage of WhatsApp compared to Instagram users..
sns.set(rc={'figure.figsize':(20,20)})
sns.catplot(y="Date", x="Usage",kind="bar",hue = "App",data=data)
```

```python
## We can see a positive correlation between usage of the phone and notifications more users used their phones when they got notifications from whatsapp
sns.lmplot(x="Usage",y="Notifications",data=data,hue="App",fit_reg=False)
plt.title("Relationship between Usage and Notifications")
plt.show()
```

```python
sns.lmplot(x="Times opened",y="Notifications",data=data,hue="App",fit_reg=False)
plt.title("Relationship between Time Usage and Notifications")
plt.show()
```
