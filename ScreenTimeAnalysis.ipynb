# Analyze the screen time of the user to find relationships between the usage of the smartphone and factors like notifications and apps used by the user.

```python
import pandas as pd
import numpy as np
import plotly.express as px
data = pd.read_csv("https://raw.githubusercontent.com/AldataSci/ScreenTimeAnalysis/main/Screentime-App-Details.csv")
data.head(15)
```

```python
print(data.head())
```

```python
data.describe()
```

```python
## Analyze the relationship between App usage and the data they were used on.. 
Fig = px.bar(data,x="Date",y="Usage",color="App",title = "App Usage")
Fig.show()
```

```python
fig = px.bar(data,x="Date",y="Times opened",color = "App",title = "Times Opened")
fig.show()
```

```python
## Create a scatterplot to demonstrate the relationship between notifcations and phone usage we use a ols regression line to show the correlation .
fig3 = px.scatter(data,x="Notifications",y="Usage",color = "App",title = "App Usage and Notifications",trendline="ols")
fig3.show()
```

```python
## Positive Correlation here where the more notifcations a user got,the more likely they were to open their phone... 
fig4 = px.scatter(data,x="Notifications",y="Times opened",color="App",title = "Times Phone Opened to Notifications",trendline = "ols")
fig4.show()
```

```python
## Users tended to use Instagram less and less as time goes on.. 
fig5 = px.line(data,x="Date",y="Usage",color = "App")
fig5.show()
```
